```python
import pandas as pd

df = pd.read_csv("./data/sale_data.csv")

print("data set:")
print(df)
```

```python
# total sale by categarory

category_sale = df.groupby("Category")["Sales"].sum()
print(category_sale)
```

```python
# avarage sale by category
avg_sale = df.groupby("Category")["Sales"].mean()
print(avg_sale)
```
